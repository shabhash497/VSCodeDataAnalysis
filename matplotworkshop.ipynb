```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data={ "Month":['jan', 'feb', 'mar', 'apr', 'may', 'jun'],
"Sales":[10000,12000,15000,13000,17000,16000],
"Profit":[2000,3000,4000,2500,3500,3000]}
df=pd.DataFrame(data)
df
```

```python
plt.figure(figsize=(8,5))
plt.plot(df['Month'], df['Sales'], color='blue', label='Sales',marker='o',linestyle='-')
plt.title('Sales Data')
plt.xlabel('Month')
plt.ylabel('Sales ($)')
plt.grid()
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(6,3))
width=0.3
plt.bar(df['Month'], df['Sales'], width=width,color='skyblue', label='Sales')
plt.bar(df['Month'], df['Profit'], width=width, color='green', label='Profit',bottom=df['Sales'])
plt.title('Sales and Profit Data by Month')
plt.xlabel('Month')
plt.grid(True)
plt.ylabel('Amount ($)')
plt.legend()
plt.tight_layout
plt.show()
```

**$Pie Chart PROFIT VS MONTH$**

```python
plt.figure(figsize=(6,3))
plt.pie(df['Profit'], autopct='%1.2f%%', labels=df['Month'], startangle=140)
plt.title('Profit Distribution by Month')
plt.show()
```

**4-Scatter Plot**

```python
plt.figure(figsize=(10,5))
plt.scatter(df['Sales'], df['Profit'], color='green', alpha=0.5,s=100,edgecolors=  'black')
plt.title('Sales vs Profit Scatter Plot')
plt.xlabel('Sales ($)')
plt.ylabel('Profit ($)')
plt.tight_layout()
plt.grid(True)
plt.show()
```

**5.Histogram**

```python
plt.figure(figsize=(8,5))
plt.hist(df['Sales'], bins=5, color='purple',edgecolor='black', alpha=0.7)
plt.title('Sales Histogram')
plt.xlabel('Sales ($)')
plt.ylabel('Frequency')
plt.tight_layout()
plt.show()
```

**6.BoxPlot**

```python
plt.figure(figsize=(10,5)) 
plt.boxplot(df['Profit'], vert=False)
plt.title('Profit Boxplot')
plt.xlabel('Profit ($)')
plt.ylabel('Frequency')
plt.tight_layout()
plt.show()
```

```python
import gradio as gr
import pandas as pd
import matplotlib.pyplot as plt
data={ "Month":['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun'],
"Sales":[10000,12000,15000,13000,17000,16000],
"Profit":[2000,3000,4000,2500,3500,3000]}
df=pd.DataFrame(data)
def genearate_plot(plot_type):
    fig= plt.figure(figsize=(10,5))
    if plot_type == 'Line Plot':
        plt.plot(df['Month'], df['Sales'], color='blue', label='Sales',marker='o',linestyle='-')
        plt.title('Sales Data')
        plt.xlabel('Month')
        plt.ylabel('Sales ($)')
        plt.grid()
        plt.legend()
    elif plot_type == 'Bar Plot':
        width=0.3
        plt.bar(df['Month'], df['Sales'], width=width,color='skyblue', label='Sales')
        plt.bar(df['Month'], df['Profit'], width=width, color='green', label='Profit',bottom=df['Sales'])
        plt.title('Sales and Profit Data by Month')
        plt.xlabel('Month')
        plt.grid(True)
        plt.ylabel('Amount ($)')
        plt.legend()
    elif plot_type == 'Pie Chart':
        plt.pie(df['Profit'], autopct='%1.2f%%', labels=df['Month'], startangle=140)
        plt.title('Profit Distribution by Month')
    elif plot_type == 'Scatter Plot':
        plt.scatter(df['Sales'], df['Profit'], color='green', alpha=0.5,s=100,edgecolors=  'black')
        plt.title('Sales vs Profit Scatter Plot')
        plt.xlabel('Sales ($)')
        plt.ylabel('Profit ($)')
    elif plot_type == 'Histogram':
        plt.hist(df['Sales'], bins=5, color='purple',edgecolor='black', alpha=0.7)
        plt.title('Sales Histogram')
        plt.xlabel('Sales ($)')
        plt.ylabel('Frequency')
    elif plot_type == 'Box Plot':
        plt.boxplot(df['Profit'], vert=False)
        plt.title('Profit Boxplot')
        plt.xlabel('Profit ($)')
        plt.ylabel('Frequency')
    plt.tight_layout()
    plt.show()
    return fig
demo = gr.Interface(fn=genearate_plot, inputs=gr.Radio(['Line Plot', 'Bar Plot', 'Pie Chart', 'Scatter Plot', 'Histogram', 'Box Plot'], label='Select Plot Type'), outputs=gr.Plot(label='Sales Data Visualization'), title='Sales & Profits Data Visualization', description='Select a plot type to visualize the sales data.')
demo.launch()
```
